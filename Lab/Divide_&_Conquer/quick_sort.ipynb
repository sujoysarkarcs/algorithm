Given an array of **n integers**, sort the array in **ascending order** using the **Quick Sort** algorithm.  
Quick Sort works by selecting a **pivot**, partitioning elements around the pivot (smaller on left, greater on right), and recursively sorting the partitions.

Where:  
- `n = number of elements`  
- Input = unsorted array  
- Output = sorted array (ascending order)

---

## <span style="color:#1f4e79;">Example 1:</span>

Input:  
arr = [9, 3, 7, 1, 6], n = 5  

Output:  
[1, 3, 6, 7, 9]

Explanation:  
Using pivot selection and partitioning, the array is gradually divided and sorted.

---

## <span style="color:#1f4e79;">Example 2:</span>

Input:  
arr = [10, 80, 30, 90, 40, 50, 70], n = 7  

Output:  
[10, 30, 40, 50, 70, 80, 90]

Explanation:  
After partitioning around pivots, each subarray gets sorted recursively.
