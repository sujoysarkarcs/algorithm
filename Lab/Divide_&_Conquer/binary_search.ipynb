Given a **sorted array** of **n integers** and a **target value**, use **Binary Search** to determine the index of the target.  
If the target exists in the array, return its index.  
If it does not exist, return **-1**.

Where:  
- `n = number of elements`  
- `target = value to search for`  
- Binary Search repeatedly divides the array into halves, reducing time complexity to **O(log n)**.

---

## <span style="color:#1f4e79;">Example 1:</span>

Input:  
arr = [1, 3, 5, 7, 9], n = 5, target = 7  

Output:  
3  

Explanation:  
7 is found at index 3 (0-based indexing).

---

## <span style="color:#1f4e79;">Example 2:</span>

Input:  
arr = [2, 4, 6, 8, 10, 12], n = 6, target = 5  

Output:  
-1  

Explanation:  
5 does not exist in the array.
