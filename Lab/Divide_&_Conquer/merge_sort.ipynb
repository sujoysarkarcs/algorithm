# <span style="color:#1f4e79;">Merge Sort â€” Divide and Conquer</span>

Given an array of **n integers**, sort the array in **ascending order** using the **Merge Sort** algorithm.  
Merge Sort works by repeatedly splitting the array into halves, sorting each half, and then merging the two sorted halves.

Where:  
- `n = number of elements`  
- Input = unsorted array  
- Output = sorted array (ascending order)

---

## <span style="color:#1f4e79;">Example 1:</span>

Input:  
arr = [5, 2, 9, 1, 3], n = 5  

Output:  
[1, 2, 3, 5, 9]

Explanation:  
The array is split into smaller halves, sorted, and merged to form the final sorted list.

---

## <span style="color:#1f4e79;">Example 2:</span>

Input:  
arr = [10, 7, 8, 4], n = 4  

Output:  
[4, 7, 8, 10]

Explanation:  
Merge Sort guarantees O(n log n) sorting regardless of the input order.
