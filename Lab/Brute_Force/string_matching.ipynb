# Brute-Force String Matching

Given a **string** and a **pattern**, determine whether the pattern appears inside the string.  
Where **n = length(string)** and **m = length(pattern)**.  
Output **"Found"** or **"Not found"**.

---

### Example 1:
string = "hello world", pattern = "world", n = 11, m = 5  
Output: **Found**

### Example 2:
string = "abcdef", pattern = "xyz", n = 6, m = 3  
Output: **Not found**
