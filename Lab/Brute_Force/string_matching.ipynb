# <span style="color:#1f4e79;">Brute-Force String Matching</span>

Given two strings **string** and **pattern**, determine whether **pattern** appears as a substring inside **string**.  
Where **n = length(string)** and **m = length(pattern)**.  
If the pattern exists, output **"Found"**; otherwise, output **"Not found"**.

---

### Example 1:
Input:  
string = "hello world", pattern = "world", n = 11, m = 5  
Output:  
Found  
Explanation: "world" occurs starting at index 6.

### Example 2:
Input:  
string = "abcdef", pattern = "xyz", n = 6, m = 3  
Output:  
Not found  
Explanation: "xyz" does not appear in "abcdef".
