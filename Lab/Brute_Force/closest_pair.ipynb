# <span style="color:#1f4e79;">Closest Pair of Points — Brute Force</span>

Given **n points** in 2D space, find the **pair of points with the minimum Euclidean distance** between them.  
Check every possible pair of points, compute their distance, and return the smallest-distance pair.

Where:  
- `n = number of points`  
- Each point = (x, y)  
- Output = the closest pair + minimum distance  

---

## <span style="color:#1f4e79;">Example 1:</span>

Input:  
points = {(0,0), (3,4), (1,1)}, n = 3  

Output:  
Minimum distance: 1.414214  
Closest pair: (0,0) — (1,1)

Explanation:  
Distance between (0,0) and (1,1) = √2 ≈ 1.414, which is the smallest.

---

## <span style="color:#1f4e79;">Example 2:</span>

Input:  
points = {(5,5), (1,1), (2,2), (10,10)}, n = 4  

Output:  
Minimum distance: 1.414214  
Closest pair: (1,1) — (2,2)

Explanation:  
Among all pair combinations, the closest pair is (1,1) and (2,2).
